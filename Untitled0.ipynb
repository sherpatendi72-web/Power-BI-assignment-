# Q1. What is Power BI and why is it used in businesses?
>> Power BI is a business intelligence (BI) and data visualization tool developed by Microsoft. It allows organizations to connect to various data sources, transform raw data into meaningful insights, and present those insights through interactive dashboards and reports. It’s available as both a desktop application (Power BI Desktop) and a cloud-based service (Power BI Service).

Businesses adopt Power BI because it helps them make data-driven decisions. Here are the main reasons:
- Data Integration
- Connects to multiple sources (Excel, SQL databases, cloud services, APIs).
- Consolidates scattered data into one platform.
- Visualization & Reporting
- Creates interactive dashboards with charts, graphs, and KPIs.
- Makes complex data easy to understand for non-technical users.
- Real-Time Insights
- Provides up-to-date information for faster decision-making.
- Enables monitoring of live business metrics.
- Collaboration
- Dashboards and reports can be shared across teams.
- Cloud-based access ensures everyone works with the same data.
- Cost & Efficiency
- Saves time by automating reporting processes.
- Reduces reliance on manual spreadsheets and static reports.

# Q3.Explain the Power BI workflow
>> The Power BI workflow can be summarized in six short steps:
- Collect Data – Connect to sources like Excel, SQL, or cloud services.
- Transform & Model – Clean, shape, and build relationships using Power Query and DAX.
- Visualize – Create interactive dashboards and reports.
- Publish – Upload reports to the Power BI Service (cloud).
- Share & Collaborate – Teams access and interact with dashboards via web or mobile.
- Refresh – Schedule automatic updates to keep data current.


```python
# Q4. List any four data cleaning tasks that can be performed in Power Query
>> Here are four common data cleaning tasks you can perform in Power Query (part of Power BI):
- Remove Duplicates – Eliminate repeated rows to ensure data accuracy.
- Filter Rows – Keep only relevant records by applying conditions (e.g., sales above a certain value).
- Split Columns – Break a single column (like “Full Name”) into multiple columns (e.g., “First Name” and “Last Name”).
- Change Data Types – Convert columns to the correct format (e.g., text to numbers, dates to proper date type).
```

# Q6.  Define Data View, Report View, and Model View. Explain the purpose of each view
>> - Data View – Shows the raw tables you’ve loaded; used to inspect and prepare data.
- Report View – The canvas for creating visuals and dashboards; used to analyze and present insights.
- Model View – Displays relationships between tables; used to define connections for accurate reporting.
